```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
Y_cos, Y_sin = np.cos(X), np.sin(X)
plt.plot(X, Y_cos);
plt.plot(X, Y_sin);
```

```python
plt.show();
```

改变显示分辨率、大小、线条粗细

```python
plt.figure(figsize=(12,9), dpi=80) # 改变分辨率和图片大小
plt.plot(X, Y_cos, linewidth=3.5, color='r') # 改变线条宽度和颜色
plt.plot(X, Y_sin, linewidth=3.5, color='b')
plt.show();
```

设置坐标轴上下限

```python
plt.xlim(-5, 5)
plt.ylim(-2, 2)
plt.show();
```

设置坐标轴记号

```python
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show();
```
